```python
#1. Menampilkan Missing Value
import pandas as pd
import numpy as np

# load dataset
df = pd.read_csv('data_tugas 2 (1).csv')

# ubah "?" menjadi NaN
df.replace("?", np.nan, inplace=True)

# tampilkan data awal
print("Data Awal:")
display(df.head())

# cek missing value
missing_table = df.isnull().sum().reset_index()
missing_table.columns = ['Variabel', 'Jumlah Missing']

print("\nJumlah Missing Value:")
display(missing_table)
```


Pada tahap ini dilakukan pembacaan dataset menggunakan pandas, kemudian ditampilkan beberapa data awal untuk melihat struktur data.

Selanjutnya dilakukan pengecekan missing value dengan mengubah simbol "?" menjadi NaN, lalu dihitung jumlah data yang hilang pada setiap variabel menggunakan fungsi `isnull().sum()`.

Hasil ditampilkan dalam bentuk tabel untuk mengetahui variabel mana yang memiliki missing value.

```python
#2. Missing Value Kategori
kategori = ['make', 'bahan-bakar', 'jumlah-pintu', 'body-style']

for col in kategori:
    df[col] = df[col].fillna("unknown")

# cek hasil
print("\nDistribusi Variabel Kategori:")
for col in kategori:
    print(f"\nDistribusi {col}:")
    print(df[col].value_counts())
```

Pada tahap ini dilakukan penanganan missing value pada variabel kategorik yaitu make, bahan-bakar, jumlah-pintu, dan body-style.

Missing value diganti dengan nilai "unknown" agar data tidak hilang dan tetap dapat digunakan dalam analisis.

Setelah itu ditampilkan distribusi tiap kategori untuk memastikan bahwa penggantian telah berhasil dilakukan.

```python
# 3. Missing Value Numerik
numerik = ['length', 'width', 'height']

# ubah ke float
for col in numerik:
    df[col] = df[col].astype(float)

# isi missing dengan mean
print("\nMean tiap variabel:")
for col in numerik:
    mean_val = df[col].mean()
    print(f"Mean {col} =", mean_val)
    df[col] = df[col].fillna(mean_val)
```

Output yang dihasilkan menampilkan nilai rata-rata (mean) dari masing-masing variabel numerik, yaitu length, width, dan height.

Nilai mean tersebut kemudian digunakan untuk mengganti data yang hilang (missing value) pada masing-masing kolom.

Dengan metode ini, data yang hilang tidak dihapus, melainkan diisi dengan nilai yang mewakili keseluruhan data, sehingga distribusi data tetap terjaga.

Setelah proses ini, tidak ada lagi missing value pada variabel numerik.

```python
#4:4: Min-Max Normalization
df['wheel-bases'] = df['wheel-bases'].astype(float)

min_val = df['wheel-bases'].min()
max_val = df['wheel-bases'].max()

df['wheel-bases-normalized'] = (df['wheel-bases'] - min_val) / (max_val - min_val)

print("\nHasil Min-Max Normalization:")
display(df[['wheel-bases', 'wheel-bases-normalized']].head())

print("Min:", df['wheel-bases-normalized'].min())
print("Max:", df['wheel-bases-normalized'].max())
```

Pada tahap ini dilakukan normalisasi pada variabel wheel-bases menggunakan metode Min-Max.

Normalisasi bertujuan untuk mengubah nilai data ke dalam rentang 0 hingga 1 dengan menggunakan rumus (x - min) / (max - min).

Hasil normalisasi disimpan dalam kolom baru bernama wheel-bases-normalized dan menunjukkan bahwa data telah berada dalam rentang yang sama.

```python
#5. Z-Score Normalization
print("\nZ-Score Normalization:")

for col in numerik:
    mean = df[col].mean()
    std = df[col].std()

    print(f"{col} -> mean: {mean}, std: {std}")
    df[col + '_zscore'] = (df[col] - mean) / std

# tampilkan hasil
display(df[['length', 'length_zscore',
            'width', 'width_zscore',
            'height', 'height_zscore']].head())

# validasi (mean harus mendekati 0)
print("\nValidasi Z-score:")
for col in ['length_zscore', 'width_zscore', 'height_zscore']:
    print(col, "mean =", df[col].mean())
```

Pada tahap ini dilakukan normalisasi menggunakan metode Z-score pada variabel length, width, dan height.

Z-score digunakan untuk mengetahui seberapa jauh suatu data dari rata-rata dalam satuan standar deviasi dengan rumus (x - mean) / std.

Hasil normalisasi menunjukkan bahwa data telah terstandarisasi dengan rata-rata mendekati 0.
